# **Inplace Operator**

```python
p=torch.rand(2,3)
q=torch.rand(2,3)
```

```python
p+q
```

```python
#p+q is stored at another location , we want to strore either in p location or q location

#store in p location
#p<--p+q
p.add_(q)
```

```python
print(p)
```

```python
#value stored in location p
```

```python
q.relu_()
```

```python
#underscore denotes inplace operations
```
