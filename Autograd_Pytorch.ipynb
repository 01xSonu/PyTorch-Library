```python
y=x**2
```

```python
z=torch.sin(y)
```

```python
x
```

```python
y
```

```python
z
```

Example

```python
#one input let cgpa:6.7 has no placement i.e X=6.7(input)  and Y =0(output)
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
w
```

```python
b
```

```python
z = w*x + b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
loss = F.binary_cross_entropy(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```

```python
y.backward()
```

```python
#after the updation is completed , no need to perform gradient tracking again and again(training time)
# option 1 - requires_grad_(False)
# option 2 - detach()
# option 3 - torch.no_grad()
```

```python
x.requires_grad_(False)
```

```python
x
```

```python
y = x ** 2
```
