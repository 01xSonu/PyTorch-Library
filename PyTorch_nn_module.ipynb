```python
 import torch
 import torch.nn as nn
```

```python
#create model
class Model(nn.Module):
  def __init__(self,num_features):

    super().__init__()
    self.linear=nn.Linear(num_features,1)
    self.sigmoid=nn.Sigmoid()

  def forward(self,features):
     out=self.linear(features)  #out(net input)=wx+b
     out=self.sigmoid(out)      #out=y_pred
     return out


```

```python
#create dataset
features=torch.rand(10,5)
```

```python
#create model
model=Model(features.shape[1])

#calling of model for forward pass
model(features)
```
